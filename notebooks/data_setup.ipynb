```python
import zipfile
import os

data_dir = '/content/drive/MyDrive/ML_final_project/data/'
os.chdir(data_dir)

zip_files = [f for f in os.listdir('.') if f.endswith('.zip')]

for zip_file in zip_files:
   with zipfile.ZipFile(zip_file, 'r') as zip_ref:
       zip_ref.extractall('.')
   os.remove(zip_file)

print("all zip files processed!")
```
